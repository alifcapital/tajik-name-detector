```python
surname_suffixes = {
    "ев",
    "ева",
    "ов",
    "ова",
    "ёв",
    "ёва",
    "цкий",
    "цкая",
    "ский",
    "ская",
    "ський",
    "ська",
    "ски",
    "ска",
    "ёвич",
    "евич",
    "ович",
    "овна",
    "евна",
    "ёвна",
    "ич",
    "ыч",
    "инична",
    "ычна",
    "енко",
    "чук",
    "юк",
    "чик",
    "чак",
    "зод",
    "зода",
}
name_suffixes = {
    "ҷон",
    "ҷан",
    "чон",
    "джон",
    "джан",
    "жон",
    "бек",
    "хон",
    "шоҳ",
    "шох",
    "шо",
    "бегим",
    "иддин",
    "уддин",
    "ддини",
    "идин",
    "удин",
    "гул",
    "улло",
    "илло",
    "туло",
    "али",
    "алӣ",
    "муҳаммад",
    "муҳамад",
    "маҳмад",
    "аҳмад",
    "махмат",
    "махмад",
    "шариф",
    "ашраф",
    "бобо",
    "дод",
    "берди",
    "кул",
    "қул",
    "хӯҷа",
    "хуҷа",
    "хуча",
    "хоҷа",
    "ходжа",
    "худжа",
    "олим",
    "нисо",
    "афруз",
    "мурод",
    "акбар",
    "мухаммад",
    "хоча",
    "бег",
    "хужа",
    "махамад",
    "гул",
    "гули",
    "мурод",
    "назар",
    "алло",
    "ҷаҳон",
    "умар",
    "уллоҳ",
    "биби",
    "ниссо",
    "мусо",
    "юсуф",
    "хусайн",
    "ҳасан",
    "мурот",
    "расул",
    "қадам",
    "хўча",
    "-Заде",
    "мухаммед",
    "berdiev",
    "уллин",
    "салим",
    "ҳоҷа",
    "-Қизи",
    "рахим",
    "карими",
    "хўҷа",
    "буби",
    "иддин",
    "бону",
    "уллина",
    "иён",
    "уллох",
    "мухамет",
    "бой",
    "саит",
    "-Мирзо",
    "-Духти",
    "-Али",
    "махамат",
    "ҳусайн",
    "улла",
    "билол",
    "сафо",
    "латиф",
    "утдин",
    "итдин",
    "-Зода",
    "ии",  # моҳбибии
}
name_suffixes = set([x + "и" for x in name_suffixes]).union(name_suffixes)

surname_suffixes_maxlen = max(len(s) for s in surname_suffixes)
name_suffixes_maxlen = max(len(s) for s in name_suffixes)


def lemmatize_name(name: str):
    name = name.title()
    for i in range(len(name) - surname_suffixes_maxlen, len(name)):
        if name[i:] in surname_suffixes:
            name = name[:i]
            break

    for i in range(max(len(name) - name_suffixes_maxlen, 3), len(name)):
        if name[i:] in name_suffixes:
            name = name[:i]
            break

    return name


for ex in positive_train_examples[1600 : 1600 + 20]:
    print(ex, end=": ")
    for word in ex.split():
        print(lemmatize_name(word), end=" ")
    print()
```

```python
names = [lemmatize_name(name) for ex in positive_train_examples for name in ex.split()]
len(names)
```

Write to file

```python
print(len(names))
names = [s for s in names if len(s) > 2]
print(len(names))
names = set(names)
print(len(names))

with open("../data/base_names.txt", "w") as f:
    f.write("\n".join(names))
```
